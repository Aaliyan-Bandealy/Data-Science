## Q2. Descriptive statistics

With the cleaned data in Q1, please provide the data summarization as below:

* Q2.1 total number of unique name, unique seller_type, unique transmission, and unique owner
* Q2.2 descriptive statistics, including the total number, mean, std, min and max regarding all km_driven
* Q2.3 descriptive statistics, including mean, std, max, and min of the number of selling_price grouped by different seller_type

```python
#Q2.1
#defining unique variables
unique_name=df_cleaned['name'].nunique()
unique_seller_type=df_cleaned['seller_type'].nunique()
unique_transmission = df_cleaned['transmission'].nunique()
unique_owner = df_cleaned['owner'].nunique()


print("Number of unique names:",unique_name)
print("Number of unique seller types:",unique_seller_type)
print("Number of unique transmissions:",unique_transmission)
print("Number of unique owners",unique_owner)
```

```python
#Q2.2
km_driven_total=df_cleaned['km_driven'].sum()
print("Total number of kms driven",km_driven_total);


km_driven_mean=df_cleaned['km_driven'].mean()
print("Mean of kms driven:",km_driven_mean);

km_driven_std=df_cleaned['km_driven'].std()
print("Std of kms driven",km_driven_std);

km_driven_min=df_cleaned['km_driven'].min()
print("Minimum number of kms driven",km_driven_min);

km_driven_max=df_cleaned['km_driven'].max()
print("Maximum number of kms driven",km_driven_max);
```

```python
#Q2.3
selling_price_grouped = df_cleaned.groupby("seller_type")["selling_price"].describe()
print("Descriptive statistics for selling_price grouped by seller_type:")
print(selling_price_grouped[["mean","std","min","max"]])
```

```python
#q2.4
years_grouped = df_cleaned.groupby("owner")["year"].describe()
print("Descriptive statistics for year grouped by owner:")
print(years_grouped[["mean","std","min","max"]])
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the cleaned dataset
df_cleaned_graph = pd.read_csv('car_sell.csv')
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='owner', y='selling_price', data=df_cleaned_graph,color = '#e0ff33')
plt.title('Selling Price by Owner type')
plt.xlabel('Owner type')
plt.ylabel('Selling Price')
```

### Observations 

#### (Owner type and Selling Price):

This box plot represents the distribution of selling price values according to owner type: First Owner, Second Owner, and so on. First Owner cars mostly include a higher median selling price compared to that of Second Owner cars, with their prices being more compact. It can thus be interpreted that First Owner cars are generally more expensive and less variable in their prices.


```python
plt.figure(figsize=(12, 6))
sns.barplot(x='transmission', y='selling_price', data=df_cleaned_graph,color = '#ff33e3')
plt.title('Selling price by type of transmission')
plt.xlabel('Transmission type')
plt.ylabel('Selling Price')
```

### Observations 

#### (Transmission type and Selling Price):

The barplot compares selling prices between manual and automatic transmission vehicles. Automatic vehicles generally have a higher median selling price, with a slightly wider range of prices.The higher prices for automatic vehicles may be attributed to their more expensive manufacturing costs and perceived convenience.

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x='seller_type', y='selling_price', data=df_cleaned_graph,color = '#09b01a')
plt.title('Selling price by seller type')
plt.xlabel('Seller Type')
plt.ylabel('Selling Price')
```

### Observations 

#### (Seller Type and Selling Price):

The box plot shows selling price by seller type: Dealer, Individual, and Trustmark Dealer. In general, the median prices for cars sold by Trustmark Dealers are the highest, with those sold by Dealers the next highest, and with those sold by Individuals generally the lowest. Again, this is because Trustmark Dealers sell certified, high-quality cars often supported by warranties, thus demanding higher prices. Individual sellers may offer less in the way of assurance and thus charge lower prices.

```python
plt.figure(figsize=(12, 6))
sns.lineplot(x='year', y='selling_price', data=df_cleaned_graph,color = '#b00925')
plt.title('Selling price by year')
plt.xlabel('Year')
plt.ylabel('Selling Price')
plt.show()
```

### Observations 

#### (Year and Selling Price):

This line plot shows the trend in selling price over the years. Selling price, has increased with each passing year. In general, the newer the car, the higher the selling price. Newer models are generally more valuable because of advances in technology and minimal wear. As cars get older, their value depreciates, leading to lower selling prices.

```python
plt.figure(figsize=(12, 6))
sns.barplot(x='fuel', y='selling_price', data=df_cleaned_graph,color = '#09a6b0')
plt.title('Selling price by Fuel Type')
plt.xlabel('Fuel')
plt.ylabel('Selling Price')
plt.show()
```

### Observations 

#### (Fuel Type and Selling Price):

This barplot compares the selling prices across different fuel types: Petrol, Diesel, etc. The median selling price for diesel cars appears to be very slightly higher than that for petrol cars, and the prices more variable, too. Diesel is more fuel-efficient and durable, so users who drive long distances could understand such a price premium built in, which explains the high prices for diesel cars.

The data reveals several patterns in the used car market, particularly in how different factors such as the number of previous owners, fuel type, transmission type, year of manufacture, and seller type influence selling prices. ## Q4. Detect and remove outliers

We define car sale records WITHOUT following ANY of the below three rules as outliers:

1. owner to be first, second and third owner only
2. Fuel to be Diesel or Petrol only
3. Seller_type to be dealer or individual only

Please remove the corresponding records in the csv file that is not qualified with the rules. After that, __print the length of the data__.

```python
#Defining Conditions
owners = ['First Owner', 'Second Owner', 'Third Owner']
fuels = ['Diesel', 'Petrol']
seller_types = ['Dealer', 'Individual']

#filter
filtered_df = df_cleaned[
    (df_cleaned['owner'].isin(owners)) &
    (df_cleaned['fuel'].isin(fuels)) &
    (df_cleaned['seller_type'].isin(seller_types))]

#saving to csv
filtered_df.to_csv('filtered_data.csv', index=False)

#length of filtered data
print("Length of filtered data:", len(filtered_df))
```
