## Task 1: Clean the datasets (10 marks)

### Step 1.1 Handle the missing values with follwoing rules (5 marks)
1. If over 50% of the values of a column are missing, the column should be removed from the data frame; 
2. For a categorical column, if a row contains a missing value, you need to delete the whole row; 
3. For a numerical column, if a row contains a missing value, you need to perform a missing value imputation with the average value of the column.

```python
#Load required libraries
import numpy as np

#check for missing values
missing = data.isnull().sum()

#finding columns with more than 50% missing values
missing_percentage = missing/len(data) * 100

#drop the missing values columns
dropped_columns = missing_percentage[missing_percentage>50].index
cleaned_data = data.drop(columns=dropped_columns)

#find categorical columns
categorical_columns = cleaned_data.select_dtypes(include=['object']).columns

#drop missing categorical values
cleaned_data = cleaned_data.dropna(subset=categorical_columns)

#missing values in numerical columns
#find numerical columns
numerical = cleaned_data.select_dtypes(include=['int64','float64']).columns

#input average values in place of missing data
cleaned_data[numerical] = cleaned_data[numerical].fillna(cleaned_data[numerical].mean())

#checking if all missing values have been removed
check = cleaned_data.isnull().sum()
print(check)

```

### Step 1.2 Handle categorical attributes (5 marks)
1. If all the categorical values of a column are unique, this column does not provide any statistical informaiton and should be deleted.
2. Use one hot encoding to convert the categorical values into numerical ones.

```python
#Check if all values are unique in categorical columns
for col in categorical_columns:
    if cleaned_data[col].nunique() == len(cleaned_data):
        cleaned_data = cleaned_data.drop(columns=[col])

#Check the remaining colums
categorical_columns = cleaned_data.select_dtypes(include=['object']).columns
print("Categorical columns left:", categorical_columns)

#One hot encoding
encoded_data = pd.get_dummies(cleaned_data, columns=categorical_columns)

# find boolean columns
bool_columns = encoded_data.select_dtypes(include=['bool']).columns

# Convert boolean values to integers
encoded_data[bool_columns] = encoded_data[bool_columns].astype(int)

# check the data
print(encoded_data[bool_columns].head())






```

## Task 2: Build a logistic regression classification model (25 marks)

### Step 2.1 Specify the features and the label, and split the dataset into training data and testing data (5 marks)
1. The attirbute "LoanApproved" is the label, which is the prediction target. The remaining attributes are the features.
2. The ratio for splitting the dataset is 80% for training and 20% for testing. Note that you need to set the "random_state" parameter as your student ID to produce your personlised splitting. Failing to do so will lose marks.

```python
#1) importing libraries
from sklearn.model_selection import train_test_split

#2) Define the features
X = encoded_data.drop('LoanApproved', axis=1)
y = encoded_data['LoanApproved']             

#splitting the dataset into train and test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=48216739)

#checking the datset
print("Training data:",  X_train.shape)
print("Testing data:", X_test.shape)
```

### Step 2.2 Build a logistic regression model (10 marks)
1. Train a logistic regression model
2. Report two classification performance metrics (accuracy and f1-score) on the testing data
3. Also report the two metrics on the training data, and compare the results with that of the testing data. Make a justification on whether the model is overfitting based on the comparison.

```python
#1) importing libraries
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, f1_score

#2) run regression model
model = LogisticRegression(max_iter=1000)

#3) train the model
model.fit(X_train, y_train)

#4) make predictions
y_train_pred = model.predict(X_train)
y_test_pred = model.predict(X_test)

#5) find accuracy and f1-score on train data
accuracy_score_train = accuracy_score(y_train, y_train_pred)
f1_score_train = f1_score(y_train, y_train_pred)

#6) find accuracy and f1-score on tesr data
accuracy_test = accuracy_score(y_test, y_test_pred)
f1_score_test = f1_score(y_test, y_test_pred)

#7) print the results
print("Training Accuracy:", accuracy_score_train * 100)
print("Training f1-Score:" ,f1_score_train)
print("Testing Accuracy:", accuracy_test * 100)
print("Testing f1-Score:", f1_score_test)
```

# Analysis

- The training accuracy would normally be far greater than the testing accuracy if a model was overfitting. Actually, the model's robustness and ability to function consistently across known and unknown data points are reinforced by the somewhat greater testing accuracy compared to the training accuracy.

### Interpretation:

### Step 2.3 Perform the recursive feature elimination (RFE) technique to identify the effective features for building the model (10 marks)
1. Visulise the change of the two performance metrics with respect to the number of eliminated features using a line chart.
2. In terms of the visualisation result, select a good value for the number of eliminated features with considering both performance maximisation and feature minimisation (two competing goals). Run the RFE again with the chosen number of eliminated features to obtain the corresponding set of retained features.

```python
#1) Import libraries
from sklearn.feature_selection import RFE
import matplotlib.pyplot as plt

#2) run regression model
model = LogisticRegression(max_iter=100) #iteration kept at this level to run code faster

#3) create lists to store information
accuracies = []
f1_scores = []
number_features = []

#4) RFE to find best features
for i in range(1, X_train.shape[1] + 1):
    best_features = RFE(estimator=model, n_features_to_select=i)
    best_features.fit(X_train, y_train)

#predictions with existing features
    y_train_pred_rfe = best_features.predict(X_train)
    y_test_pred_rfe = best_features.predict(X_test)
    
#calculate performance metrics score for the test data
    accuracy = accuracy_score(y_test, y_test_pred_rfe)
    f1 = f1_score(y_test, y_test_pred_rfe)
    
#append the lists
    accuracies.append(accuracy)
    f1_scores.append(f1)
    number_features.append(i)

#5) visualize the data
plt.figure(figsize=(12,8))
plt.plot(number_features, accuracies, label='Accuracy')
plt.plot(number_features, f1_scores, label='F1 Score')
plt.xlabel('Retained Features')
plt.ylabel('Performance Scores')
plt.title('Performance Scores vs. Retained Features')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#rerun RFE with optimal number of reatined features
features = 20

optimal_rfe = RFE(estimator=model, n_features_to_select=features)
optimal_rfe.fit(X_train, y_train)

#get list of the features
features_list = X_train.columns[optimal_rfe.support_]
print("Optimal features from RFE:" , features_list)
```

## Task 3: Build a KNN classification model (25 marks)

### Step 3.1 Build 1-NN classifier (5 marks)
1. Slect the features identifed in Step 2.3 for this task
2. Buid 1-NN classifier and report two classification performance metrics (accuracy and f1-score) on the testing data
3. Also report the two metrics on the training data, and compare the results with that of the testing data. Make a justification on whether the model is overfitting based on the comparison.

```python
#1) Importing libraries
from sklearn.neighbors import KNeighborsClassifier

#2) Use the selected features list
X_train_selected = X_train[features_list]
X_test_selected = X_test[features_list]

#3) Creating the 1-NN classifier
model_knn = KNeighborsClassifier(n_neighbors=1)

#4) Training the model
model_knn.fit(X_train_selected, y_train)

#5) Making predictions
y_train_pred_knn = model_knn.predict(X_train_selected)
y_test_pred_knn = model_knn.predict(X_test_selected)

#6) Check the model on training data
train_accuracy_knn = accuracy_score(y_train, y_train_pred_knn)
train_f1_score_knn = f1_score(y_train, y_train_pred_knn)

#7) Check the model on testing data
test_accuracy_knn = accuracy_score(y_test, y_test_pred_knn)
test_f1_score_knn = f1_score(y_test, y_test_pred_knn)

# Print the performance metrics
print("Training Accuracy:" ,train_accuracy_knn * 100)
print("Training F1-Score:" ,train_f1_score_knn)
print("Testing Accuracy:" ,test_accuracy_knn * 100)
print("Testing F1-Score:" ,test_f1_score_knn)
```

# Analysis

## 1. Training Accuracy and F1-Score
- The model appears to have learned the training data, as evidenced by its **100% accuracy** and **F1-Score of 1.00** on the training set. 
- 1-NN (1-Nearest Neighbor) classifiers are prone to perfectly fitting the training data by nature because they classify based on the one nearest occurrence.

## 2. Testing Accuracy and F1-Score
- The testing data's **accuracy of 82.79%** and **F1-Score of 0.65** indicate a discernible drop in performance, suggesting the classifier's generalization abilities are lacking. The F1-score of 0.65, significantly lower than the ideal 1.00 during training, shows a lack of balance between precision and recall.
- The model is overfitting, as evidenced by the wide discrepancy between testing performance (**82.79%**) and training performance (**100%**). When faced with new, unknown data, the classifier performs poorly, despite excelling on the training data. This conclusion is further supported by the **F1-score disparity** (1.00 vs. 0.65).

## Justification
- Since each new occurrence is given the label of its closest match in the training set, the classifier essentially memorizes the training data. This behavior leads to **overfitting**, where the model learns specific patterns or noise in the training set rather than capturing broader trends that would apply to unseen examples.


### Step 3.2 Use the grid search and cross validation techniques to study the performance change with respect to the hyperparameter K (10 marks)
1. User grid search to search K in the range (1, 30) both inclusive with 5-fold cross validation. The performance metric used for search is accuracy.

```python
#1) Importing libraries
from sklearn.model_selection import GridSearchCV, cross_val_score

#2) Define the parameter grid for K
param_grid = {'n_neighbors': range(1, 31)}

#3) Do the grid search
search_grid = GridSearchCV(estimator=model_knn, param_grid=param_grid, cv=5, scoring='accuracy')
search_grid.fit(X_train_selected, y_train)

#4) Get the results of the grid search
results = search_grid.cv_results_

# Calculate the mean of each K
mean_test_scores = results['mean_test_score']
k_values = list(param_grid['n_neighbors'])

#5) Visualize
plt.figure(figsize=(12,8))
plt.plot(k_values, mean_test_scores, marker='o', label='Mean Test')
plt.xlabel('Number of Neighbors-K')
plt.ylabel('Mean Test')
plt.title('Mean Test vs. Number of Neighbors-K')
plt.grid(True)
plt.legend()
plt.show()
```
